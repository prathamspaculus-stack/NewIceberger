```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_calls = pd.read_csv("2023 to 2026(Holiday).csv")
```

```python
daily_calls
```

```python
daily = daily_calls.copy()
```

```python
daily["date"] = pd.to_datetime(daily["date"], format="%d-%m-%Y")
```

```python
daily['date'].dt.day
```

```python
daily['dayofweek'] = daily['date'].dt.dayofweek
daily['month'] = daily['date'].dt.month
daily['weekofyear'] = daily['date'].dt.isocalendar().week.astype(int)
```

```python
daily.head()
```

```python
daily = daily.sort_values('date').reset_index(drop=True)
```

```python
daily['lag_1'] = daily["call_count"].shift(1)
daily['lag_7'] = daily["call_count"].shift(7)
daily['lag_14'] = daily["call_count"].shift(14)
daily['lag_21'] = daily["call_count"].shift(21)
daily['lag_28'] = daily["call_count"].shift(28)
```

```python
daily['roll_mean_7'] = daily['call_count'].rolling(7).mean()
daily['roll_std_7'] = daily['call_count'].rolling(7).std()
daily['roll_mean_14'] = daily['call_count'].rolling(14).mean()
daily['roll_std_14'] = daily['call_count'].rolling(14).std()
```

```python
daily
```

```python
daily.corr()
```

```python
train = daily[daily['date'] < "2025-07-01"]
test = daily[daily['date'] >= '2025-07-01']
```

```python
features = [
    "lag_1","lag_7","lag_21","lag_28","roll_mean_7","roll_mean_14","dayofweek","roll_std_7"
]
```

```python
target = "call_count"
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    objective="count:poisson",
    n_estimators=380,
    learning_rate=0.06,
    max_depth=2,
    subsample=0.5,
    colsample_bytree=0.9,
    random_state=42
)
```

```python
model.fit(train[features], train[target])
```

```python
preds = model.predict(test[features])
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(test[target], preds)
print("MAE:", mae)
```

```python
# remove weekday rows where call_count < 200
df = daily[~((daily['dayofweek'] <= 4) & (daily['call_count'] < 200))]

# sort by date
df = df.sort_values('date')

df
```

```python
train = df[df['date'] < "2025-07-01"]
test = df[df['date'] >= '2025-07-01']
```

```python
model.fit(train[features], train[target])
```

```python
preds = model.predict(test[features])
```
