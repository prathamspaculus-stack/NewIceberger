```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_calls = pd.read_csv("2023 to 2026(Holiday).csv")
```

```python
daily_calls
```

```python
plt.figure()
plt.plot(daily_calls['date'], daily_calls['call_count'], color = 'blue')

plt.fill_between(daily_calls['date'], daily_calls['call_count'], color = 'green', alpha = 0.6)

plt.title('Daily Call Count Volume')
plt.xlabel('Date')
plt.ylabel('Number of Calls')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Only Week days Call Volume

```python
import pandas as pd
import matplotlib.pyplot as plt

# convert date column
daily_calls['date'] = pd.to_datetime(daily_calls['date'], format='%d-%m-%Y')

# keep only weekdays (Mon-Fri)
df = daily_calls[daily_calls['date'].dt.dayofweek < 5]

# sort by date (important for trend graph)
df = df.sort_values('date')
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['date'], df['call_count'], color='blue')
plt.fill_between(df['date'], df['call_count'], color='green', alpha=0.6)

plt.title("Week Days Call Volume")
plt.xlabel("date")
plt.ylabel("Number of calls")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# filter dates where call count is less than 150
low_calls = df[df['call_count'] < 200]

# show the result
print(low_calls[['date','call_count']])
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['date'], df['call_count'], color='blue')

# highlight low call dates
low_calls = df[df['call_count'] < 150]
plt.scatter(low_calls['date'], low_calls['call_count'], color='red')

plt.title("Week Days Call Volume")
plt.xlabel("date")
plt.ylabel("Number of calls")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
