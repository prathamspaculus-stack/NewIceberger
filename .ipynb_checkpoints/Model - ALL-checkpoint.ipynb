```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_calls = pd.read_csv("2023 to 2026(all).csv")
```

```python
daily_calls
```

```python
daily_calls = daily_calls.drop(columns = "Unnamed: 0")
```

```python
daily_calls.head()
```

```python
daily = daily_calls.copy()
```

```python
daily["date"] = pd.to_datetime(daily["date"])
```

```python
daily["date"].dt.day
```

```python
daily['dayofweek'] = daily['date'].dt.dayofweek
daily['month'] = daily['date'].dt.month
daily['weekofyear'] = daily['date'].dt.isocalendar().week.astype(int)
```

```python
daily.head()
```

```python
daily = daily.sort_values('date').reset_index(drop=True)
```

```python
daily['lag_1'] = daily["call_count"].shift(1)
daily['lag_7'] = daily['call_count'].shift(7)
daily['lag_14'] = daily['call_count'].shift(14)
daily['lag_21'] = daily['call_count'].shift(21)
daily['lag_28'] = daily['call_count'].shift(28)
```

```python
daily['roll_mean_7'] = daily['call_count'].rolling(7).mean()
daily['roll_std_7']= daily['call_count'].rolling(7).std()

daily['roll_mean_14']= daily['call_count'].rolling(14).mean()
daily['roll_std_14'] = daily['call_count'].rolling(14).std()
```

```python
daily
```

```python
daily.corr()
```

```python
train = daily[daily["date"] < "2025-07-1"]
test = daily[daily["date"] >= '2025-07-1']
```

```python
features = [
    "lag_1","lag_7","lag_21","lag_28","roll_mean_7","roll_mean_14","dayofweek"
]
```

```python
target = "call_count"
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    objective="count:poisson",
    n_estimators=310,
    learning_rate=0.06,
    max_depth=2,
    subsample=0.8,
    colsample_bytree=0.9,
    random_state=42
)
```

```python
model.fit(train[features], train[target])
```

```python
preds = model.predict(test[features])
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(test[target], preds)
print("MAE:", mae)
```

```python
daily["call_count"].mean()
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
```

```python
model = XGBRegressor(
    objective="count:poisson",
    random_state=42,
)
tscv = TimeSeriesSplit(n_splits=3)
```

```python
param_grid = {
    "max_depth": [2, 3, 4, 5],
    "learning_rate": [0.03, 0.05,0.06, 0.07, 0.1],
    "n_estimators": [100, 200, 300,320,325, 400],
    "subsample": [0.7, 0.8, 0.9],
    "colsample_bytree": [0.7, 0.8, 0.9]
}
```

```python
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,   
    cv=tscv,
    scoring="neg_mean_squared_error",  
    n_jobs=-1,            
)

grid_search.fit(train[features], train[target])
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)

best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(test[features])
```

```python
mae = mean_absolute_error(test[target], y_pred)
print("MAE:", mae)
```
